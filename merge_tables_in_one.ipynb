### Fill missing values 

```python
from geopy.geocoders import Nominatim

locator = Nominatim(user_agent="myGeocoder")

City = []

for street in restaurant['street']:


    if locator.geocode(f"{street} SF", timeout=10000) is None:
        City.append("New York")

    else:
        City.append('San Francisco')

    for l in street:
        if l=='-':
            City.pop()
            City.append('San Francisco')

del restaurant['city']

restaurant.insert(1,'city', City)

restaurant
```

## Split order_creation_date to date, time, and day

```python
import datetime
from datetime import date
import calendar

order_full_table_copy = order_full_table_copy.join(order_full_table_copy["order_creation_date"].str.split(" ", expand=True))
order_full_table_copy[0] = order_full_table_copy[0].apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%y"))
order_full_table_copy[0] = order_full_table_copy[0].apply(lambda x: calendar.day_name[x.weekday()])
order_full_table_copy = order_full_table_copy.rename(columns={0:"order_day", 1:"order_time"})

order_full_table_copy = order_full_table_copy.join(order_full_table_copy["order_creation_date"].str.split(" ", expand=True))
order_full_table_copy = order_full_table_copy.drop(['order_creation_date', 1], axis=1)
order_full_table_copy = order_full_table_copy.rename(columns={0:'order_creation_date'})

order_full_table_copy
```

```python
order_full_table_copy.isnull().sum().sum()
```

## Add period column

```python
# Period

periods = {
    "Brunch": (0, 1200),
    "Lunch": (1200, 1500),
    "Tea": (1500, 1800),
    "Dinner": (1800, 2359)
}


def get_period(time):
    time = time.split(":")
    time = int(time[0] + time[1])
    for period, (period_start, period_end) in periods.items():
        if period_start <= time <= period_end:
            return period


# To apply the function to the "order_time" column and create the "period" column
order_full_table_copy["period"] = order_full_table_copy["order_time"].apply(get_period)
```

## Add season column

```python
# Seasons


"""
The first is necessary to convert the "order_time_creation" content in datetime objects. 
The whole process uses only Pandas methods (to_datetime and, later, date_range) so there is no need
to import the Datetime library. 
"""

def season_of_date(date):
    year = str(date.year)
    # This dictionary must absolutely stay inside the function so the "year" is not out of scope
    seasons = {
        'spring': pd.date_range(start = year + '-03-21 00:00:00', end = year + '-06-20 00:00:00'),
        'summer': pd.date_range(start = year + '-06-21 00:00:00', end = year + '-09-22 00:00:00'),
        'autumn': pd.date_range(start = year + '-09-23 00:00:00', end = year + '-12-20 00:00:00')
    }
    if date in seasons['spring']:
        return 'spring'
    if date in seasons['summer']:
        return 'summer'
    if date in seasons['autumn']:
        return 'autumn'
    else:
        return 'winter'
    
order_full_table_copy["order_creation_date"] = pd.to_datetime(order_full_table_copy["order_creation_date"])
order_full_table_copy["order_season"] = order_full_table_copy["order_creation_date"].apply(season_of_date)
```
